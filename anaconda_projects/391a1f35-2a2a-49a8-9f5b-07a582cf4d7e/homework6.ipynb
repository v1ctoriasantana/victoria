```python
import numpy as np
# --- 3.1 ------
lst = [20, 27, 9, 12, 31, 17, 4, 50, 39, 35]
arr = np.array(lst)
print("Original array:", arr)

print("Times 4:", arr * 4)

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard deviation:", np.std(arr))

imin = np.argmin(arr)
imax = np.argmax(arr)
print("Index of min:", imin, "value:", arr[imin])
print("Index of max:", imax, "value:", arr[imax])

print("Min value:", np.min(arr))
print("Max value:", np.max(arr))

greater_than_20 = arr[arr > 20]
print("Values > 20:", greater_than_20)

# ---- 3.2 ----
matrix = np.array([
    [5, 12, 17, 9],
    [8, 21, 14, 3],
    [6, 13, 19, 11]
])

print("Matrix:\n", matrix)

print("Shape:", matrix.shape)  # (3,4)

print("Third column:", matrix[:, 2])

print("Second row:", matrix[1, :])

print("Matrix * 2:\n", matrix * 2)

print("Column means:", np.mean(matrix, axis=0))

print("Overall mean:", np.mean(matrix))

print("Row maxima:", np.max(matrix, axis=1))
print("Column minima:", np.min(matrix, axis=0))

flat = matrix.flatten()
print("Flattened array:", flat)
print("New shape:", flat.shape)

imin = np.argmin(flat)
imax = np.argmax(flat)
print("Flat index min:", imin, "Flat index max:", imax)

print("Min coords:", np.unravel_index(imin, matrix.shape))
print("Max coords:", np.unravel_index(imax, matrix.shape))

gt15 = matrix[matrix > 15]
print("Values > 15:", gt15)

print("Sorted by row:\n", np.sort(matrix, axis=1))

# --- 4 -----
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def prime_rows(arr):
    mask = [any(is_prime(x) for x in row) for row in arr]
    return arr[mask]

arr = np.array([
    [2, 3, 5],
    [14, 6, 8],
    [11, 13, 17],
    [7, 10, 13]
])

print("Rows with primes:\n", prime_rows(arr))

# --- 5.1-5.3 ----
def checkerboard():
    board = np.zeros((8, 8), dtype=int)
    board[::2, ::2] = 1  # fill even rows, even cols
    board[1::2, 1::2] = 1  # fill odd rows, odd cols
    return board

print(checkerboard())

# ---- 5.4 ----
def reverse_checkerboard():
    board = np.zeros((8, 8), dtype=int)
    board[::2, 1::2] = 1
    board[1::2, ::2] = 1
    return board

print(reverse_checkerboard())

# --- 6 ----
def expansion(arr, n):
    def expand_word(word):
        return (' ' * n).join(list(word))
    return np.array([expand_word(w) for w in arr])

universe = np.array(['galaxy', 'clusters'])
print(expansion(universe, 1))
print(expansion(universe, 2))

# ---- 7 -----
def secondDimmest(arr):
    sorted_rows = np.sort(arr, axis=1)
    return sorted_rows[:, 1]

np.random.seed(123)
stars = np.random.randint(500, 2000, (5, 5))
print(stars)
print("Second-brightest per row:", secondDimmest(stars))
```
