```python


import numpy as np
import matplotlib.pyplot as plt

# ## Function Definition
# Define a function to make a sine wave given x, amplitude A, and angular frequency .

def make_sine_wave(x, A, w):
    return A * np.sin(w * x)

# ## Create the x array
# Generate 1000 evenly spaced values from 0 to 2.

x = np.linspace(0, 2 * np.pi, 1000)

# ## Amplitude and Frequency Arrays
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]

# ## Create Subplots
fig, axs = plt.subplots(1, 3, figsize=(25, 5))

linestyles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]  # variety of line styles

# ## Plot 1: Varying Amplitude
for A, style in zip(amplitude, linestyles):
    y = make_sine_wave(x, A, 1)
    axs[0].plot(x, y, linestyle=style, label=f"A={A}, =1")

# ## Plot 2: Varying Frequency
for w, style in zip(frequency, linestyles):
    y = make_sine_wave(x, 1, w)
    axs[1].plot(x, y, linestyle=style, label=f"A=1, ={w}")

# ## Plot 3: Varying Amplitude and Frequency
for A, w, style in zip(amplitude, frequency, linestyles):
    y = make_sine_wave(x, A, w)
    axs[2].plot(x, y, linestyle=style, label=f"A={A}, ={w}")

# ## Customize the plots
titles = ['Varying Amplitude', 'Varying Frequency', 'Varying Both']
for i, ax in enumerate(axs):
    ax.set_title(titles[i], fontsize=16)
    ax.set_xlabel('x', fontsize=14)
    ax.set_ylabel('y', fontsize=14)
    ax.legend(fontsize=12)

plt.show()
```
