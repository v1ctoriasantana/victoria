```python

import numpy as np
import matplotlib.pyplot as plt

# ## Import Data
# Load star data using NumPy.
# File path: python_decal_fa25/victoria/course_assignments/homework7/star_data.txt
data = np.loadtxt("../course_assignments/homework7/star_data.txt", skiprows=1)
print("Data shape:", data.shape)

# ## Separate Columns
# 1st column: Magnitude (brightness)
# 2nd column: Temperature (Kelvin)
magnitude = data[:, 0]
temperature = data[:, 1]

# ## Explore Data
print("Magnitude min:", np.min(magnitude))
print("Magnitude max:", np.max(magnitude))
print("Temperature min:", np.min(temperature))
print("Temperature max:", np.max(temperature))

# ## Identify Brightest/Dimmest/Hottest/Coldest Stars
# Brightest = lowest magnitude
# Dimmest = highest magnitude
# Hottest = highest temperature
# Coldest = lowest temperature
brightest_idx = np.argmin(magnitude)
dimmest_idx = np.argmax(magnitude)
hottest_idx = np.argmax(temperature)
coldest_idx = np.argmin(temperature)

# ## Convert to Log Scale
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

# ## Plot HR Diagram
plt.figure(figsize=(8, 6))
plt.scatter(log_temperature, log_magnitude, color='skyblue', s=30, label='Stars')

# Highlight special stars
plt.scatter(log_temperature[brightest_idx], log_magnitude[brightest_idx],
            color='black', marker='*', s=300, label='Brightest')
plt.scatter(log_temperature[dimmest_idx], log_magnitude[dimmest_idx],
            color='purple', marker='*', s=300, label='Dimmest')
plt.scatter(log_temperature[hottest_idx], log_magnitude[hottest_idx],
            color='green', marker='*', s=300, label='Hottest')
plt.scatter(log_temperature[coldest_idx], log_magnitude[coldest_idx],
            color='orange', marker='*', s=300, label='Coldest')

# Invert both axes (astronomy convention)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.title("HertzsprungRussell Diagram", fontsize=16)
plt.xlabel("Temperature (log K)", fontsize=14)
plt.ylabel("Magnitude", fontsize=14)
plt.legend()
plt.show()
```
